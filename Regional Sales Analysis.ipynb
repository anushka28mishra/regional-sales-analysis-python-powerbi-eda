```python
#Merge with Customers
df = df_sales.merge(
    df_customers,
    how = 'left',
    left_on = 'Customer Name Index',
    right_on = 'Customer Index'
)
```

```python
df.head()
```

```python
#Merge with Products
df = df.merge(
    df_products,
    how = 'left',
    left_on = 'Product Description Index',
    right_on = 'Index'
)
```

```python
df.head()
```

```python
#Merge with Regions
df = df.merge(
    df_regions,
    how = 'left',
    left_on = 'Delivery Region Index',
    right_on = 'id'
)
```

```python
df.head()
```

```python
#Merge with State Regions
df = df.merge(
    df_state_reg[["State Code", "Region"]],
    how = 'left',
    left_on = 'state_code',
    right_on = 'State Code'
)
```

```python
df.head()
```

```python
#Merge with Budgets
df = df.merge(
    df_budgets,
    how = 'left',
    on = 'Product Name'
)
```

```python
#clean up redundant columns 
cols_to_drop = ['Customer Index', 'Index', 'id', 'State Code']
df = df.drop(columns = cols_to_drop, errors = 'ignore')

df.head()
```

```python
#convert all columns to lower case for consistency and ease
df.columns = df.columns.str.lower()
df.columns.values
```

```python
df.to_csv('file.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure order_date is in datetime format
df['order_date'] = pd.to_datetime(df['order_date'])

# Group by month and sum the revenue (you can change 'revenue' to another metric if needed)
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['revenue'].sum().reset_index()

# Convert the period back to a datetime for plotting
monthly_sales['order_date'] = monthly_sales['order_date'].dt.to_timestamp()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['order_date'], monthly_sales['revenue'], marker='o')

# Chart styling
plt.title('Monthly Sales Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Ensure order_date is a datetime column
df['order_date'] = pd.to_datetime(df['order_date'])

# Filter out records from Jan and Feb 2018
df_new = df[~((df['order_date'].dt.year == 2018) & (df['order_date'].dt.month.isin([1, 2])))]
```

```python
# Aggregate total revenue by product
product_revenue = (
    df_new.groupby('product_name')['revenue']
      .sum()
      .reset_index()
      .sort_values(by='revenue', ascending=False)
      .head(10)
)

# Plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=product_revenue, x='product_name', y='revenue', palette='tab10')

# Add formatted dollar labels on top of each bar
for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f"${v:,.0f}" for v in container.datavalues],  # adds commas + $
        label_type='edge',
        padding=3,
        fontsize=10,
        color='black'
    )

# Chart styling
ax.set_title('Top 10 Products by Revenue', fontsize=14)
ax.set_xlabel('Product', fontsize=12)
ax.set_ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Unit Price Distribution per Product
product_revenue = df.groupby('product_name')['revenue'].sum().reset_index()
top_10_products = product_revenue.sort_values(by = 'revenue', ascending = False).head(10)['product_name'].tolist()
df_top_10 = df[df['product_name'].isin(top_10_products)]
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top_10, x='product_name', y='unit-price', palette='tab10')

# Chart sty
plt.title('Unit Price Distribution per Product', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Unit Price ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# --- Aggregate data ---
state_summary = (
    df.groupby('state_name')
      .agg({
          'revenue': 'sum',
          'order_number': 'count'  # assuming order_number is unique per order
      })
      .reset_index()
      .rename(columns={'order_number': 'order_count'})
)

# --- Get top 10 states by revenue ---
top10_revenue = state_summary.sort_values(by='revenue', ascending=False).head(10)

# --- Get top 10 states by order count ---
top10_orders = state_summary.sort_values(by='order_count', ascending=False).head(10)

# --- Plot setup ---
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- Plot 1: Top 10 States by Revenue ---
sns.barplot(data=top10_revenue, x='state_name', y='revenue', palette='tab10', ax=axes[0])
axes[0].set_title('Top 10 States by Revenue', fontsize=14)
axes[0].set_xlabel('State', fontsize=12)
axes[0].set_ylabel('Total Revenue ($)', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True, axis='y', linestyle='--', alpha=0.7)

# Add dollar labels
for container in axes[0].containers:
    axes[0].bar_label(container, labels=[f"${v:,.0f}" for v in container.datavalues], padding=3, fontsize=9)

# --- Plot 2: Top 10 States by Order Count ---
sns.barplot(data=top10_orders, x='state_name', y='order_count', palette='tab10', ax=axes[1])
axes[1].set_title('Top 10 States by Order Count', fontsize=14)
axes[1].set_xlabel('State', fontsize=12)
axes[1].set_ylabel('Order Count', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(True, axis='y', linestyle='--', alpha=0.7)

# Add count labels
for container in axes[1].containers:
    axes[1].bar_label(container, labels=[f"{int(v):,}" for v in container.datavalues], padding=3, fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# --- Compute average profit margin by channel ---
channel_profit = (
    df.groupby('channel')['profit_margin_pct']
      .mean()
      .reset_index()
      .sort_values(by='profit_margin_pct', ascending=False)
)

# --- Plot ---
plt.figure(figsize=(8, 6))
ax = sns.barplot(data=channel_profit, x='channel', y='profit_margin_pct', palette='tab10')

# --- Add percentage labels on top of bars ---
for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f"{v:,.1f}%" for v in container.datavalues],  # format as percentage
        label_type='edge',
        padding=3,
        fontsize=10,
        color='black'
    )

# --- Style ---
ax.set_title('Average Profit Margin by Channel', fontsize=14)
ax.set_xlabel('Sales Channel', fontsize=12)
ax.set_ylabel('Average Profit Margin (%)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by customer and calculate total revenue
customer_revenue = df.groupby('customer_name', as_index=False)['revenue'].sum()

# Sort by revenue
customer_revenue = customer_revenue.sort_values(by='revenue', ascending=False)

# Top 10 and bottom 10 customers
top_10 = customer_revenue.head(10)
bottom_10 = customer_revenue.tail(10)

# Create subplots
plt.figure(figsize=(14, 6))

# Top 10 customers bar chart
plt.subplot(1, 2, 1)
plt.barh(top_10['customer_name'], top_10['revenue'], color='green')
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Customer')
plt.gca().invert_yaxis()  # Highest revenue on top

# Bottom 10 customers bar chart
plt.subplot(1, 2, 2)
plt.barh(bottom_10['customer_name'], bottom_10['revenue'], color='red')
plt.title('Bottom 10 Customers by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Customer')
plt.gca().invert_yaxis()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 1: Aggregate data per customer ---
customer_summary = df.groupby('customer_name', as_index=False).agg({
    'revenue': 'sum',
    'profit_margin_pct': 'mean'  # or 'sum' depending on how you want to treat margin
})

# --- Step 2: Create scatter plot ---
plt.figure(figsize=(10, 6))
plt.scatter(
    customer_summary['revenue'],
    customer_summary['profit_margin_pct'],
    c=customer_summary['profit_margin_pct'],
    cmap='viridis',
    s=80,
    alpha=0.7,
    edgecolor='k'
)

# --- Step 3: Add labels and formatting ---
plt.title('Customer Segmentation: Revenue vs Profit Margin (%)', fontsize=14)
plt.xlabel('Total Revenue', fontsize=12)
plt.ylabel('Average Profit Margin (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.colorbar(label='Profit Margin (%)')

# Optional: annotate top customers for context
top_customers = customer_summary.sort_values(by='revenue', ascending=False).head(3)
for i, row in top_customers.iterrows():
    plt.text(row['revenue'], row['profit_margin_pct'] + 0.5, row['customer_name'],
             fontsize=9, weight='bold', color='darkblue')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 1: Select only desired numeric columns ---
selected_cols = ['revenue', 'profit', 'order_quantity', 'unit-price']  # adjust column name if needed
subset_df = df[selected_cols]

# --- Step 2: Compute correlation matrix ---
corr = subset_df.corr()

# --- Step 3: Plot heatmap ---
plt.figure(figsize=(7, 6))
sns.heatmap(
    corr,
    annot=True,           # show values
    fmt=".2f",            # 2 decimal places
    cmap='coolwarm',      # color scheme
    center=0,
    linewidths=0.5,
    square=True
)

plt.title('Correlation Heatmap: Revenue, Profit, Quantity, Unit Price', fontsize=14, pad=15)
plt.tight_layout()
plt.show()

          
```
